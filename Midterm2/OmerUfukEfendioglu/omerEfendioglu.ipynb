```python
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 
```

```python
data[['temp', 'area']]
```

```python
print(data['area'].min())
print(data['area'].max())

print(data['temp'].min())
print(data['temp'].max())

print(data['wind'].min())
print(data['wind'].max())

print(data['rain'].min())
print(data['rain'].max())


print(data['RH'].min())
print(data['RH'].max())
```

```python
def total_area_group (row):
    if row['area'] >= 0.0 and row['area'] <= 100.0 :
        return 1
    if row['area'] > 100.0 and row['area'] <= 200.0 :
        return 2
    if row['area'] > 200.0 and row['area'] <= 300.0 :
        return 3
    if row['area'] > 300.0 and row['area'] <= 400.0 :
        return 4
    if row['area'] > 400.0 and row['area'] <= 500.0 :
        return 5
    if row['area'] > 500.0 and row['area'] <= 600.0 :
        return 6
    if row['area'] > 700.0 and row['area'] <= 800.0 :
        return 7
    if row['area'] > 800.0 and row['area'] <= 900.0 :
        return 8
    if row['area'] > 900.0 and row['area'] <= 1000.0 :
        return 9
    if row['area'] > 1000.0 and row['area'] <= 1100.0 :
        return 10
```

```python
def total_temp_group (row):
    if row['temp'] >= 0.0 and row['temp'] <= 2.2 :
        return 1
    if row['temp'] > 2.2 and row['temp'] <= 5.5 :
        return 2
    if row['temp'] > 5.5 and row['temp'] <= 8.8 :
        return 3
    if row['temp'] > 8.8 and row['temp'] <= 12.1 :
        return 4
    if row['temp'] > 12.1 and row['temp'] <= 15.4 :
        return 5
    if row['temp'] > 15.4 and row['temp'] <= 18.7 :
        return 6
    if row['temp'] > 18.7 and row['temp'] <= 22.0 :
        return 7
    if row['temp'] > 22.0 and row['temp'] <= 25.3 :
        return 8
    if row['temp'] > 25.3 and row['temp'] <= 31.9 :
        return 9
    if row['temp'] > 31.9 and row['temp'] <= 33.3 :
        return 10
```

```python
def total_wind_group (row):
    if row['wind'] >= 0.0 and row['wind'] <= 0.4 :
        return 1
    if row['wind'] > 0.4 and row['wind'] <= 1.4 :
        return 2
    if row['wind'] > 1.4 and row['wind'] <= 2.4 :
        return 3
    if row['wind'] > 2.4 and row['wind'] <= 3.4 :
        return 4
    if row['wind'] > 3.4 and row['wind'] <= 4.4 :
        return 5
    if row['wind'] > 4.4 and row['wind'] <= 5.4 :
        return 6
    if row['wind'] > 5.4 and row['wind'] <= 6.4 :
        return 7
    if row['wind'] > 6.4 and row['wind'] <= 7.4 :
        return 8
    if row['wind'] > 7.4 and row['wind'] <= 8.4 :
        return 9
    if row['wind'] > 8.4 and row['wind'] <= 9.4 :
        return 10
```

```python
def total_rain_group (row):
    if row['rain'] >= 0.0 and row['rain'] <= 1.4 :
        return 1
    if row['rain'] > 1.4 and row['rain'] <= 2.4 :
        return 2
    if row['rain'] > 2.4 and row['rain'] <= 3.4 :
        return 3
    if row['rain'] > 3.4 and row['rain'] <= 4.4 :
        return 4
    if row['rain'] > 4.4 and row['rain'] <= 5.4 :
        return 5
    if row['rain'] > 5.4 and row['rain'] <= 6.4 :
        return 6
```

```python
def total_RH_group (row):
    if row['RH'] >= 0.0 and row['RH'] <= 15.0 :
        return 1
    if row['RH'] > 15.0 and row['RH'] <= 35.0 :
        return 2
    if row['RH'] > 35.0 and row['RH'] <= 55.0 :
        return 3
    if row['RH'] > 55.0 and row['RH'] <= 75.0 :
        return 4
    if row['RH'] > 75.0 and row['RH'] <= 100.0 :
        return 5
```

```python
data['area_group'] = data.apply (lambda row: total_area_group (row), axis =1)
data['temp_group'] = data.apply (lambda row: total_temp_group (row), axis =1)
data['wind_group'] = data.apply (lambda row: total_wind_group (row), axis =1)
data['rain_group'] = data.apply (lambda row: total_rain_group (row), axis =1)
data['RH_group'] = data.apply (lambda row: total_RH_group (row), axis =1)
```

```python
sns.factorplot(x= 'temp_group', y= 'area', data= data, kind= "bar", ci=None)
plt.xlabel('Temp')
plt.ylabel('Area')
plt.show()
```

```python
model = smf.ols(formula='area ~ C(temp_group)', data=data)
results = model.fit()
results.summary()
```

```python
mc = multi.MultiComparison(data['area'], data['temp_group'])
res = mc.tukeyhsd()
res.summary()
```

```python
sns.factorplot(x= 'wind_group', y= 'area', data= data, kind= "bar", ci=None)
plt.xlabel('Wind')
plt.ylabel('Area')
plt.show()
```

```python
model1 = smf.ols(formula='area ~ C(wind_group)', data=data)
results1 = model1.fit()
results1.summary()
```

```python
mc1 = multi.MultiComparison(data['area'], data['wind_group'])
res1 = mc1.tukeyhsd()
res1.summary()
```

```python
sns.factorplot(x= 'rain_group', y= 'area', data= data, kind= "bar", ci=None)
plt.xlabel('Rain')
plt.ylabel('Area')
plt.show()
```

```python
model1 = smf.ols(formula='area ~ C(rain_group)', data=data)
results1 = model1.fit()
results1.summary()
```

```python
mc1 = multi.MultiComparison(data['area'], data['rain_group'])
res1 = mc1.tukeyhsd()
res1.summary()
```
