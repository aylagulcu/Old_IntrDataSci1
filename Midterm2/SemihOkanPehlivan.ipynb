```python
def CO20or1 (row):
    if row['CO2_all'] > 550000  :
        return 0
    else:
        return 1
    
dfa['CO20or1']= dfa.apply(lambda row : CO20or1(row), axis= 1)
```

```python
### [527954.73008560168, 550000, 600000, 650000, 700000, 750000, 800000, 850000, 900000, 946091.56523296749]
print(dfa['Agri_all'].min())
print(dfa['Agri_all'].max())
def AgriculturalGroup (row):
    if row['Agri_all'] > 3.0 and row['Agri_all']<= 100000 :
        return 1
    if row['Agri_all'] > 100000.0 and row['Agri_all']<= 10001000.0 :
        return 2
    if row['Agri_all'] > 1000000.0 and row['Agri_all']<= 5000000.0 :
        return 3
    if row['Agri_all'] > 5000000.0 and row['Agri_all']<= 10000000.0 :
        return 4
    if row['Agri_all'] > 10000000.0 and row['Agri_all']<= 20000000.0 :
        return 5
    if row['Agri_all'] > 20000000.0 and row['Agri_all']<= 40000000.0 :
        return 6
    if row['Agri_all'] > 40000000.0 and row['Agri_all']<= 49430849.0 :
        return 7
    
```

```python
dfa['AgriCAT'] = dfa.apply (lambda row: AgriculturalGroup (row),axis=1)
dfa = dfa.dropna()
print(dfa)
```
